```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv('IPL Matches 2008-2020.csv')
matches.head() #dl_applied -> rain affected
```

```python
delivery = pd.read_csv('deliveries.csv')
```

```python

delivery.head()
```

```python
 delivery.query("inning == 1 and batting_team == 'Kolkata Knight Riders' and bowling_team == 'Royal Challengers Bangalore'").head()
```

```python
delivery.shape
```

```python
total_score_df = delivery.groupby(['id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df
```

```python

total_score_df = total_score_df[total_score_df['inning'] == 1]
```

```python
total_score_df
```

```python
match_df = matches.merge(total_score_df[['id','total_runs']],left_on='id',right_on='id')
```

```python
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[['id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(delivery,on='id')
```

```python

delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.head()
```

```python
delivery_df['current_score'] = delivery_df.groupby('id').cumsum()['total_runs_y']
```

```python
delivery_df.tail()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
wickets = delivery_df.groupby('id').cumsum()['player_dismissed'].values
delivery_df['wickets'] = 10 - wickets
```

```python
#current run rate = runs/ overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#required run rate
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.sample()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.isnull().sum()
```

```python
final_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0] # rrr results in inf, 1/0
```

```python
final_df['crr'].max()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df.describe()
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
X_train.describe()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python

pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[10]
```
